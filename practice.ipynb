```python
#extract text from pdf file
import io

from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage


def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos = set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages,
                                  password=password,
                                  caching=caching,
                                  check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
import spacy
import nltk
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
nlp=spacy.load('en')

```

```python
#preprocessing
from spacy.lang.en.stop_words import STOP_WORDS
#lematization & remove stopwords
#remove five from stopwords and add Fund to stopwords
STOP_WORDS.add("Fund")
STOP_WORDS.remove("five")

nlp = spacy.load('en_core_web_sm')
doc = nlp(raw_text)
lemmas = [token.lemma_ for token in doc]
a_lemmas = [lemma for lemma in lemmas  if lemma not in STOP_WORDS]
clean_text=(' '.join(a_lemmas))
```

```python
#split text into sentences

sent_tokenize_list = sent_tokenize(clean_text)
len(sent_tokenize_list)
sent_tokenize_list

#split each sentence into words
word_tokenize_list=([word_tokenize(x) for x in sent_tokenize_list])
word_tokenize_list
len(word_tokenize_list)
word_tokenize_list
```

```python
#using spacy to extract NER
nlp=spacy.load('en')
doc=nlp(clean_text)
dic=dict([(str(x), x.label_) for x in doc.ents])
keys=[key for key in dic]
keys


```

```python
#filter element that only contains NER

result=[x for x in word_tokenize_list for x1 in keys if x1 in x]

result=[' '.join(x) for x in result]

#remove duplicates
clean_result=[]
for i in result:
    if i not in clean_result:
        clean_result.append(i)


clean_result
```
